## Modelo

```python
print(list(df.columns))
```

```python
df[['ScheduledDay','AppointmentDay']].head()
```

```python
df['fecha_reserva'] = df['ScheduledDay'].str[0:10]
df['fecha_cita'] = df['AppointmentDay'].str[0:10]
```

```python
df[['fecha_reserva','fecha_cita']].head()
```

```python
df['fecha_reserva'] = pd.to_datetime(df['fecha_reserva'],format='%Y-%m-%d')
```

```python
df['fecha_cita'] = pd.to_datetime(df['fecha_cita'],format='%Y-%m-%d')
```

```python
df['dias_entre'] = df['fecha_cita'] - df['fecha_reserva']
```

```python
import numpy as np
df['dias_entre_1'] = df['dias_entre'] / np.timedelta64(1, 'D')
```

```python
df['Gender'] = df['Gender'].map({'F':0,'M':1}).astype(int)
```

```python
df['Neighbourhood'].value_counts()
```

```python
df['N_jc'] = 0
df['N_jc'][df['Neighbourhood']=='JARDIM CAMBURI'] = 1
```

```python
df['N_mo'] = 0
df['N_mo'][df['Neighbourhood']=='MARIA ORTIZ'] = 1
```

```python
df['No-show'] = df['No-show'].map({'No':0,'Yes':1}).astype(int)
```

```python
df['No-show'].value_counts()
```

```python
features = ['Gender','Age','Scholarship', 'Hipertension', 'Diabetes', 'Handcap', 'SMS_received','dias_entre_1','N_jc','N_mo']
target = ['No-show']
```

```python
for var in features:
    print(var,df[var].dtype)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
train, test = train_test_split(df, test_size = 0.2)
```

```python
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

```python
#clf = tree.DecisionTreeClassifier()
clf = RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```python
confusion_matrix(y_train,clf.predict(X_train))
```

```python
recall_score(y_train,clf.predict(X_train))
```

```python
confusion_matrix(y_test,clf.predict(X_test))
```

```python
accuracy_score(y_test,clf.predict(X_test))
```

```python
recall_score(y_test,clf.predict(X_test))
```
